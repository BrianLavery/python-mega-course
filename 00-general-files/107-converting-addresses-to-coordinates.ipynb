```python
import pandas
# import geopy
# dir(geopy) # => use to see methods
from geopy.geocoders import ArcGIS
nom = ArcGIS()
```

```python
df = pandas.read_json('../files/supermarkets/supermarkets.json')
df
```

```python
# This is how we invoke geocoding
# If not found this returns a None object
n = nom.geocode("20 Derryvore Lane, Portadown, Northern Ireland, BT63 5RS", timeout = None)
```

```python
n.latitude
```

```python
n.longitude
```

```python
# We want to update address to be full address we can send to geocode
df["Address"] = df["Address"] + ", " + df["City"] + ", " + df["State"] + ", " + df["Country"]
df
```

```python
# We won't iterate over rows here to get coordinates for each. Use pands.apply
df["Coordinates"] = df["Address"].apply(nom.geocode)
df
```

```python
# Can check one value
df.iloc[0,7]
```

```python
# Another way
df.Coordinates[0]
```

```python
df.Coordinates[0].latitude
```

```python
# To set latitude as a column we use a lambda
df["Latitude"] = df["Coordinates"].apply(lambda x: x.latitude)
df
```

```python
# Can apply some if logic in case we get None values
# Can also add longitude
df["Latitude"] = df["Coordinates"].apply(lambda x: x.latitude if x != None else None)
df["Longitude"] = df["Coordinates"].apply(lambda x: x.longitude if x != None else None)
df
```
