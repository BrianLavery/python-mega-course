```python
import pandas
```

```python
df = pandas.read_json('./files/supermarkets/supermarkets.json')
df
```

```python
df = df.set_index("Address")
df
```

```python
# We are specifying a range using labels
df.loc["735 Dolores St":"332 Hill St", "ID":"Country"]
```

```python
# Can access single cells using same approach
df.loc["735 Dolores St", "Name"]
```

```python
# To get entire column/ row
df.loc[:, "Name"]
```

```python
# Can convert data set to a list with built in function
list(df.loc[:, "Name"])
```

```python
df
```

```python
# More commonly use indices not labels to access sets
# use iloc for this
# First specify rows then columns (same as mathematical matrices)
# Numbers are upper bound exclusive
# With labels the upper bound is included
df.iloc[1:3, 0:3]
```

```python
df.iloc[4, :]
```
