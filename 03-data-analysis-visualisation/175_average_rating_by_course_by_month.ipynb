### Overview of the dataframe

```python
import pandas
from datetime import datetime # Need to compare times
from pytz import utc # Need to compare times
import matplotlib.pyplot as plt # plt is naming convention

# We use parse dates argument here to get Timestamps as datetime objects not strings
data = pandas.read_csv("./review_analysis/reviews.csv", parse_dates = ['Timestamp'])
data.head()
```

```python
data['Month'] = data['Timestamp'].dt.strftime('%Y-%m')

# We want to preserve courses in this data frame produced
month_average_crs_stacked = data.groupby(['Month', 'Course Name']).mean()
month_average_crs_stacked # Produces multi-level index
```

```python
# We unstack to avoid a multi-level index
month_average_crs = data.groupby(['Month', 'Course Name']).mean().unstack()
month_average_crs # Produces something like a pivot table
```

```python
# Simplest way to plot
month_average_crs.plot(figsize = (25, 8))
```

## Count example

```python
# When we use count the lable is a mess as we aren't just getting data series for ratings but get for all data series
month_average_crs_count = data.groupby(['Month', 'Course Name']).count().unstack()
month_average_crs_count.plot(figsize = (25, 8))
```

## Count example fixed

```python
# Group by returns a data frame so we extract only the rating
month_average_crs_count_rating = data.groupby(['Month', 'Course Name'])['Rating'].count().unstack()
month_average_crs_count_rating.plot(figsize = (25, 8))
```
