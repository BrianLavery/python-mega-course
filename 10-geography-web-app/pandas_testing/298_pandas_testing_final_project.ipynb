```python
import pandas
from geopy.geocoders import ArcGIS
nom = ArcGIS()

df_address = pandas.read_csv("./supermarkets_address.csv")
df_no_address = pandas.read_csv("./supermarkets_no_address.csv")
```

```python
df_address['Latitude'] = [1,2,3,4,5,6]
df_address['Longitude'] = [1,2,3,4,5,6]
```

```python
lats = []
longs = []

for index, row in df_address.iterrows():
    coord = nom.geocode(row["Address"], timeout = None)
    if coord == None:
        lats.append(None)
        lats.append(None)
    else:
        lat = coord.latitude
        long = coord.longitude
        lats.append(lat)
        longs.append(long)
```

```python
df_address['Latitude'] = lats
df_address['Longitude'] = longs
```

```python
df_address.drop('new', inplace=True, axis=1)
df_address.drop('Longitue', inplace=True, axis=1)
```

```python
df_address
```

```python
df_address["Coordinates"] = df_address["Address"].apply(nom.geocode)
```

```python
df_address.rename(columns={'Address': 'address'}, inplace = True)
```
